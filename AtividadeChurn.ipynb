```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Iniciando nossa leitura dos dados
df = pd.read_csv('CHURN_TELECON_MOD08_TAREFA.csv', delimiter=';')

df.head(10)
```

```python
# Verificar o tipo de cada coluna
df.dtypes
```

```python
# Verificando quantos valores nulos há em cada coluna
df.isnull().sum()
```

```python
# Mostrar a porcentagem de cada dado nulo nas colunas
(df.isnull().sum() / len(df)) * 100
```

```python
# Removendo os dados vazios da coluna 'Churn', há apenas 5 dados
df.dropna(subset=['Churn'], inplace=True)
```

```python
# Verificando se deu certo a exclusão
df['Churn'].isnull().sum()
```

```python
# Deixando as letras das colunas em caixa alta
df['Servico_Internet'] = df['Servico_Internet'].str.upper()
```

```python
# Padronizando a letra dos dados em caixa alta
df['Servico_Internet'] = df['Servico_Internet'].replace('dsl', 'DSL')
```

```python
df.head(10)
```

```python
# Removendo colunas que não considero uma informação útil
df.drop(columns=['Genero'], inplace=True)
df.drop(columns=['Idoso'], inplace=True)
df.drop(columns=['Dependentes'], inplace=True)
df.drop(columns=['Casado'], inplace=True)
```

```python
# Padronizando as colunas do nosso DataFrame em PT-BR
df.rename(columns={'ID_CLiente': 'ID_Cliente'}, inplace=True)
df.rename(columns={'Dependents': 'Dependentes'}, inplace=True)
df.rename(columns={'Internet_Service': 'Servico_Internet'}, inplace=True)
df.rename(columns={'PhoneService': 'Servico_Telefonico'}, inplace=True)
df.rename(columns={'PaymentMethod': 'Metodo_Pagamento'}, inplace=True)
df.rename(columns={'StreamingTV': 'TV_por_Assinatura'}, inplace=True)
```

```python
df.head(10)
```

```python
# Verificando se há outros valores nas colunas
print(df['Metodo_Pagamento'].unique())
print(df['Suporte_Tecnico'].unique())
print(df['TV_por_Assinatura'].unique())
```

```python
# Algumas verificações para garantir a exclusividade dos dados. "\n\n" usado para pular duas linhas entre os códigos
valores_nulos = df["Servico_Seguranca"].value_counts()
print(valores_nulos, end="\n\n")

valores_nulos = df["Metodo_Pagamento"].value_counts()
print(valores_nulos, end="\n\n")

valores_nulos = df["TV_por_Assinatura"].value_counts()
print(valores_nulos, end="\n\n")

valores_nulos = df["Suporte_Tecnico"].value_counts()
print(valores_nulos, end="\n\n")
```

```python
# Verificando a coluna que possui uma "pequena" quantidade de outliers, porém, muito importante, então faremos o seguinte tratamento:
df['Pagamento_Mensal'].mean()
```

```python
# Gráfico para analisar se possui algum outlier na coluna "Pagamento_Mensal"
sns.set_style('whitegrid')

# Plota o boxplot da coluna Pagamento_Mensal
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='Pagamento_Mensal', color='skyblue')
plt.title('Boxplot da Coluna Pagamento_Mensal')
plt.ylabel('Pagamento Mensal')
plt.show()
```

Não possui nenhum outlier, analisando bem, a média seria uma abordagem mais segura para preencher os dados vazios da coluna analisada, não poderíamos excluí-la por ser uma coluna importante.

```python
# Removendo os valores nulos e preenchendo pela média
df["Pagamento_Mensal"] = df["Pagamento_Mensal"].fillna(df["Pagamento_Mensal"].mean())
```

```python
# Verificando quantos dados nulos há na coluna "Servico_Telefonico"
valores_nulos = df["Servico_Telefonico"].isnull().sum()
print(valores_nulos)
```

```python
# Preenchendo os valores nulos pelo "Não Informado"
df['Servico_Telefonico'].fillna('Não Informado', inplace=True)
```

Bom, como não sei se a pessoa possui ou não o serviço telefônico, decidi por criar uma nova categoria "Não Informado", não quis apagar a coluna por ser,  aparentemente relevante.
